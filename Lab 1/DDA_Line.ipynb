```python
import math
from graphics import *
win = GraphWin("daa",1000, 1000)
```

```python
x1=int(input("enter x1 "))
y1=int(input("enter y1 "))
x2=int(input("enter x2 "))
y2=int(input("enter y2 "))

dx=x2-x1
dy=y2-y1

if dx>dy:
    step=dx
else:
    step=dy
```

```python
Xinc=float(dx/step)
Yinc=float(dy/step)
x=x1
y=y1
print( 'x= ', round(x),"y= ", round(y))
```

```python
for i in range(step):
    x += Xinc
    y += Yinc
    print( 'x= ', round(x),"y= ", round(y))
    pt = Point(x, y)
    pt.draw(win)
```
